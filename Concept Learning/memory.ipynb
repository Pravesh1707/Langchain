```python
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()
```

```python
message_hist = ChatMessageHistory()
```

```python
message_hist.add_user_message('Hello')
message_hist.add_ai_message('Hi There')
print(message_hist)
```

```python
store={}

def get_history(session_id : str):
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
```

```python
prompt = ChatPromptTemplate.from_messages([('system','You are an Helpful AI Asssistant'),('human','{input}')])
```

```python
llm = ChatOpenAI(model='gpt-3.5-turbo')
parser = StrOutputParser()
chain = prompt | llm | parser
```

```python
chat_with_history = RunnableWithMessageHistory(chain,get_history, input_messages_key='input',output_messages_key='output')
```

```python
response = chat_with_history.invoke({'input':'Hello i am Rahul'},config={'configurable':{'session_id' : 'Rahul6163'}})
```
