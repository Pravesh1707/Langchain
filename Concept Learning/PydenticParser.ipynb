## Pydentic Output Parser

```python
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
from pydantic import BaseModel,Field
from typing import Optional,Literal
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
class Data(BaseModel):
    Name : str = Field(description='Provide me the Name of the Product')
    Age : int= Field(gt =18,description='Return the Age of the person')
    Gender : Literal['Male','Female'] = Field(description='mention the Sentiment of the user Review')
    City : str = Field(description='city should be from Uttar Pradesh only')
    State : str = Field(description='State should be Uttar Pradesh only') 
```

```python
model = ChatGoogleGenerativeAI(model='gemini-3.5-flash')
```

```python
parser = PydanticOutputParser(pydantic_object=Data)
```

```python
prompt = PromptTemplate(template='Provide the Name, Age,Gender,City,State of a fictional Character from {nationality} in {format}',
                        input_variables=['nationality'],partial_variables={'format':parser.get_format_instructions})
```

```python
final_prompt = prompt.invoke({'nationality':'India'})
```

```python
response = model.invoke(final_prompt)
print(response.content)
```

```python
from langchain_openai import ChatOpenAI
model1 = ChatOpenAI(model='gpt-3.5-turbo')
```

```python
chain = prompt | model1 | parser
```

```python
result = chain.invoke({'nationality':'India'})
```

```python
print(result)
```
