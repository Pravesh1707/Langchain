## How to get Structured Output from LLMs

```python
# !pip install pydantic 
```

```python
from pydantic import BaseModel,Field
from typing import Optional,Literal
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
Review = """I recently bought the noise cancellation headphones. 500 and I have been using them daily for the past two weeks. 
        The sound is absolutely fantastic with the base and clear vehicles. Noise noise cancellation works very well, specially on 
        flights and the battery easily lost more than 18 hours on a single charge, however, I found them a bit uncomfortable after 
        viewing them for more than three hours constantly as a year cups gets slightly warm. Also, the companion mobile app is not 
        very intuitive. Overall I would rate them four point 4.5 out of 5 because of the amazing performance and low long battery 
        life, but with room for improvement in comfort and design."""
```

```python
# class Data(BaseModel):
#     Product : str ='No Info'
#     Pros : list[str]
#     Cons : list[str]
#     Review : Optional[str]=None
#     Ratings : Field(gt=0,lt=5,default=3,description='Customer')
```

```python
class Data(BaseModel):
    Product : str = Field(description='Provide me the Name of the Product')
    Pros : list[str] = Field(description='Mention Positive Points mentioned in the ')
    Cons : list[str]
    Sentiment : Literal['Positive','Negative'] = Field(description='mention the Sentiment of the user Review')
    Review : Optional[str]=None
    Ratings : float = Field(gt=1,lt=5,default=3,description='Rating Provided by the user as per review')
```

```python
load_dotenv()
```

```python
model = ChatOpenAI(model='gpt-3.5-turbo')
```

```python
str_model = model.with_structured_output(Data)
```

```python
response = str_model.invoke(Review)
```

```python
response.dict()
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
import streamlit as st 
```

```python
reportLLM = ChatGoogleGenerativeAI(model='gemini-3.5-flash')
```

```python
summary = ChatOpenAI(model='gpt-3.5-turbo')
```

```python
prompt1 = PromptTemplate(template='Generate a Detailed Report on the {topic}', input_variables=['topic'])
prompt2 = PromptTemplate(template='Create a 5 point Summary on {Report}', input_variables=['Report'])
```

```python
topic = input('Enter the Topic')
```

```python
topic
```

```python
final1 = prompt1.invoke({'topic':topic})
```

```python
parser = StrOutputParser()
```

```python
chain = prompt1 | reportLLM | parser | summary | parser
```

response = chain.invoke(topic)

```python
response = chain.invoke({'topic':topic})
```
