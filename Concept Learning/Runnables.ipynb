# Runnables

```python
from langchain_openai import ChatOpenAI,OpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableSequence, RunnableLambda,RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser, PydanticOutputParser
from dotenv import load_dotenv
```

```python
load_dotenv()
```

### Runnable Sequence

```python
prompt = PromptTemplate(template='Generate a Joke on the {topic}',input_variables=['topic'])
```

```python
model = ChatOpenAI(model='gpt-4o')
```

```python
parser = StrOutputParser()
```

```python
chain = RunnableSequence(prompt, model, parser)
```

```python
result = chain.invoke({'topic':'Narander Modi'})
```
