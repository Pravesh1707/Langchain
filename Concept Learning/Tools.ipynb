### Custom Tool using Pydantic

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class AdditionInput(BaseModel):
    a : int = Field(description='First number to add')
    b : int = Field(description='Secound number to add')
```

```python
def add(a:int,b:int)->int:
    return a+b
```

```python
addition_tool = StructuredTool.from_function(
    func=add,name='Addition Tool',description='Tool helps in Adding two numbers',args_schema=AdditionInput
)
```

```python
result = addition_tool.invoke({'a':7,'b':4})
```

```python
result
```

```python
print(addition_tool.name, '\n', addition_tool.description)
```

## ToolKit
#### it is a collection or Bundel of Related tools that serves a commen purpose (packaged Together)
#### eg : GoogleDriveToolKit --> GoogleDriveCreateFile,GoogleDriveSearchTool, GoogleDrivesReadFile

```python
from langchain.tools import tool
```

```python
@tool('addition')
def add(a:int,b:int)->int:
    '''Helps Adding two Numbers'''
    return a+b

@tool('multiplication')
def multiply(a:int,b:int)->int:
    '''Helps Multiplying two Numbers'''
    return a*b
```

```python
class MathsToolKit:

    def get_Tools(self):
        return [add,multiply]
```

```python
toolkit = MathsToolKit()
tools = toolkit.get_Tools()
```

```python
for tool in tools:
    print(tool.name, '-->', tool.description)
```

## Tool binding
#### it is a step where we register our Tools with a LLM, so that LLM is aware of which tools are available and what each tool does and inputs required via structure

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_core.messages import HumanMessage
from langchain.tools import tool
```

```python
load_dotenv()
```

```python
llm = ChatOpenAI(model='gpt-4o')
```

```python
@tool('addition')
def add(a:int,b:int)->int:
    '''Helps Adding two Numbers'''
    return a+b

@tool('multiplication')
def multiply(a:int,b:int)->int:
    '''Helps Multiplying two Numbers'''
    return a*b
```

```python
tool_binding = llm.bind_tools([add,multiply])
```

```python
response = tool_binding.invoke([HumanMessage(content='multiply 5 and 10')])
```

```python
response.tool_calls
```

## Tool Execution
#### 

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_core.messages import HumanMessage,AIMessage,ToolMessage
from langchain.tools import tool

load_dotenv()


@tool('addition')
def add(a:int,b:int)->int:
    '''Helps Adding two Numbers'''
    return a+b

@tool('multiplication')
def multiply(a:int,b:int)->int:
    '''Helps Multiplying two Numbers'''
    return a*b

llm = ChatOpenAI(model='gpt-4o')

message =[]

human_msg = HumanMessage(content='multiply 4 and 5')
message.append(human_msg)

llm_with_tool = llm.bind_tools([add,multiply])
AI_msg = llm_with_tool.invoke(message)

message.append(AI_msg)
tool_call = AI_msg.tool_calls[0]

tool_msg = ToolMessage(content=str(tool_call),name=tool_call['name'],tool_call_id=tool_call['id'])
message.append(tool_msg)

final_result = llm_with_tool.invoke(message)
```

```python
print(final_result.content)
```
