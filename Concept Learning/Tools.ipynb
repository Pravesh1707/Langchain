### 2. Custom Tool - user defined tools
### Requirements :

###     2. Typing thinting - input output typr
###     3. Decorator - 

### 1. Name
### 2. Description
### 3. args

#### LLM looks to the JSON Schema of the tool to run the required tool


```python
from langchain.tools import tool
```

```python
@tool('Addition')
def add(a:int,b:int)->int:
    "This Function helps in adding two numbers that are provided"
    return a+b
```

```python
result = add.invoke({'a':2,'b':4})
print(result)
```

```python
print(add.name ,add.description)
```

### Custom Tool using Pydantic

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class AdditionInput(BaseModel):
    a : int = Field(description='First number to add')
    b : int = Field(description='Secound number to add')
```

```python
def add(a:int,b:int)->int:
    return a+b
```

```python
addition_tool = StructuredTool.from_function(
    func=add,name='Addition Tool',description='Tool helps in Adding two numbers',args_schema=AdditionInput
)
```

```python
result = addition_tool.invoke({'a':7,'b':4})
```

```python
result
```

```python
print(addition_tool.name, '\n', addition_tool.description)
```

## ToolKit
#### it is a collection or Bundel of Related tools that serves a commen purpose (packaged Together)
#### eg : GoogleDriveToolKit --> GoogleDriveCreateFile,GoogleDriveSearchTool, GoogleDrivesReadFile

```python
from langchain.tools import tool
```

```python
@tool('addition')
def add(a:int,b:int)->int:
    '''Helps Adding two Numbers'''
    return a+b

@tool('multiplication')
def multiply(a:int,b:int)->int:
    '''Helps Multiplying two Numbers'''
    return a*b
```

```python
class MathsToolKit:

    def get_Tools(self):
        return [add,multiply]
```

```python
toolkit = MathsToolKit()
tools = toolkit.get_Tools()
```

```python
for tool in tools:
    print(tool.name, '-->', tool.description)
```
