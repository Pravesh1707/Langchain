## Parallel Chains

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
prompt_1 = PromptTemplate(template='Generate a Summary on this {Topic}', input_variables=['Topic'])
prompt_2 = PromptTemplate(template='Generate 5 Quize Questionon this {Topic}', input_variables=['Topic'])
```

```python
gemini = ChatGoogleGenerativeAI(model='gemini-3.1-flash-lite')
gpt = ChatOpenAI(model='gpt-4o')
```

```python
Parser = StrOutputParser()
```

```python
parallel_chain = RunnableParallel({'summary':prompt_1 | gemini | Parser, 'quiz': prompt_2 | gpt | Parser})
```

```python
prompt_3 = PromptTemplate(template=' Merge the following {summary} and {quiz} together', input_variables=['quiz','summary'])
```

```python
chain = prompt_3 | gemini | Parser
```

```python
final_chain = parallel_chain | chain
```

```python
response = final_chain.invoke({'Topic': 'Generative AI'})
```
