## Task 11 : Unified Preprocessing Pipeline

````python
from langchain_community.document_loaders import TextLoader,PyMuPDFLoader,CSVLoader,DirectoryLoader
from langchain_text_splitters import CharacterTextSplitter
````

````python
def load_and_split(path_or_file):
    loader = PyMuPDFLoader(path_or_file)
    doc = loader.load()
    splitter = CharacterTextSplitter(chunk_size=300, chunk_overlap =30)
    chunks = splitter.split_documents(doc)
    return chunks
````

````python
result = load_and_split("./Data/Sample-filled-in-MR.pdf")
````

````python
result
````

## Task 12: Observations & Insights

````python
# TextLoader --> Text files
# CSVLoader --> CSV Files
# PyPDF,PyMuPDFLoader --> PDF Files
# DirectoryLoader --> Directory of files
````

````python
# Small Text  --> CharacterTextSplitter
# Large PDFs  --> RecursiveCharacterTextSplitter
# web Data  --> RecursiveCharacterTextSplitter
````

````python
# Chunk Overlap is importtant to preserve the context of the last chunk and the Current chunk. 
# It helps in maintaining the continuity of information when splitting text into smaller segments. 
# By overlapping chunks, you ensure that important context from the end of one chunk is carried over to the beginning of the next chunk, which can be crucial for understanding and processing the text effectively.
````
